```python
import numpy as np
import matplotlib 
from pylab import *
```

```python
figure()
t = array([10,11,12,13])
y = array([20,30,40,50])
plot(t,y,'+',t,2*y,'D-.') #plot takes 1st parameter as X and second as Y
hold(False)
show()
#savefig('line.png')
#import os
#[file for file in os.listdir('.') if file.endswith('png')]
```

```python
I = arange(6)
figure()
#hold(True)
plot(I,sin(I),'o',lw = 3, ms=8)
plot(I,cos(I),'--',lw=2,ms=5)
show()
```

```python
R = 1.2
I = arange(0,4*pi,0.01)
plot(cos(I)*R,sin(I)*R)
grid(True)
axhline(color='gray')
axvline(color='gray')
axis('auto')
title(r'$\frac{\pi}{5}$')
xticks([-1,0,1],('Negative','Neutral','Positive'))
yticks(arange(-1.5,2.0,1))
show()

#close(1)
```
