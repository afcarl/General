```python
dis = ['DISEASES OF HEART', 'MALIGNANT NEOPLASMS',
       'INFLUENZA AND PNEUMONIA', 'CHRONIC LOWER RESPIRATORY DISEASES',
       'CEREBROVASCULAR DISEASE', u'DIABETES MELLITUS',
       'ACCIDENTS EXCEPT DRUG POISONING',]
db2009 = [59409,39183,6714,4533,4296,5034,2949]
db2010 = [53061,39597,7326,5076,5061,4704,2739]
db2011 = [50166,40008,7389,5274,5220,5205,3006]
```

```python
len(dis)
l = range(7)
l1 = map(lambda x: x -0.2, l)
l2 = map(lambda x: x +0.2, l)
c = '#5F9EA0'
```

```python
ax = plt.subplot(111)
ax.bar(l1,db2009,width=0.2,color=c,align='center',alpha = 1)
ax.bar(l, db2010,width=0.2,color=c,align='center',alpha = 0.6)
ax.bar(l2, db2011,width=0.2,color=c, align='center',alpha = 0.3)
plt.ylabel('Count')
plt.xlabel('Diseases')
plt.xticks(l, dis, rotation='vertical')
plt.title('Most common diseases for 2009, 2010, 2011')
plt.show()
```
